```python
# Developing the histogram for the dataset using Seaborn
plt.figure(figsize=(10, 6))
sns.histplot(data=df, x='prod_inc', bins=10, kde=True, color='skyblue')
plt.title('Psychological Effects of COVID-19 (Seaborn)')
plt.xlabel('Rate of productivity increase (a psycological effect)')
plt.ylabel('Count or Frequency')
plt.show()
```


***HISTOGRAM***

X-Axis (Horizontal Axis):

Represents the variable being measured or observed
It is divided into intervals or bins, each representing a specific range of values.
Here, we have taken ***prod_inc*** column of dataframe as X-axis which means the rate of productivity increase

Y-Axis (Vertical Axis):

Represents the frequency or count of observations falling into each bin.
The height of each bar corresponds to the number of data points in that bin.

***Analysis:-***

The rate of productivity is symmetric according to the histogram.
When the productivity rate increases the count of people falling in to that group also increases but gradually it decreases when productivity reaches .50 or the median value.

```python
#colum "relaxed"- Rating of how relaxed the person is feeling
# Create a histogram
df_t=df.head(500)
fig = px.histogram(df_t, x='relaxed', nbins=10, title='Psycological effect to relaxation after covid19')
# Update layout to add title
fig.update_layout(title_text='Psycological effect to relaxation after covid19', xaxis_title='How much relaxed',yaxis_title='count of people', xaxis_tickformat='%')

# Update traces to add legend
fig.update_traces(marker_line_color='black', marker_line_width=1, opacity=0.7, name='Category')
# Add a shape to highlight the 0th position in red
#fig.add_trace(px.scatter(df, x='relaxed').data[0])
for value in df_t['relaxed']:
    fig.add_shape(type='line', x0=value, x1=value, y0=0, y1=100, line=dict(color='red', width=1))
# Show the plot
fig.show()
```

**Analysis**

Out of a sample of 500 individuals, 62 respondents reported being completely relaxed (100% relaxation), while 74 individuals described their state as similar to feeling miserable. According to the analysis, the histogram, which represents a range from -1 to +1, indicates that a considerable number of people fall into the -1 category.

The majority of respondents, representing a significant portion, reported being not at all relaxed (0% relaxation) after experiencing the effects of COVID-19. On the other hand, approximately 120 individuals reported a level of relaxation at 50%

```python
import matplotlib.pyplot as plt
from pywaffle import Waffle
```

```python
df_t = df.head(500)

# Create a Waffle chart
fig = plt.figure(
    FigureClass=Waffle,
    rows=10,  # Adjust the number of rows as needed
    values=dict(df_t['relaxed'].value_counts()),  # Use the 'relaxed' column for values
    legend={'loc': 'upper left', 'bbox_to_anchor': (1, 1)},
    title={'label': 'Distribution of relaxation after COVID-19(rated -1 to 1)', 'loc': 'center'},
)

# Show the plot
plt.show()
```


A Waffle chart represents data using a grid of small squares. Each tile typically represents a unit, percentage, or quantity of the whole. It conveys distribution or proportion of categorical data, in this scenario it shows distribution of relaxation level of people.


***Analysis***

The largest group of participants, indicated by the turquoise-green tiles, reported experiencing no relaxation (0% relaxation) after being affected by COVID-19, constituting a substantial proportion. Conversely, the second-largest group, represented by orange tiles, indicated a relaxation level of 50%.

Within this dataset, the light green tile set signifies individuals who are entirely relaxed, comprising the smallest segment of the population.

The set of pink-colored tiles represents individuals who are extremely unrelaxed or disturbed.
